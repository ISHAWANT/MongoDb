```python
# find the record at once present in the collection with specific field value. 
name_record = collection.find({},{'Name'}) 

for index , record in enumerate(name_record):
    print(f"Record {index}: {record}")
    print("========================================================") 
```

```python
# let's delete one record from the name filed. 

deleted_record = collection.delete_one({'Name': 'Rohit'}) 
```

```python
# let's print the all data after deleting the record. 
find_all_records = collection.find() 
for index, record in enumerate(find_all_records):
    print(f"Record {index}: {record}")
    print("========================================================") 
```

```python
# Let's delete multiple records from the collection. 

deleted_multiple_records = collection.delete_many({'Age': {'$gte': 30}})   

deleted_multiple_records.deleted_count 
```

```python
# let's print all records after deleting the multiple records. 
find_all_records = collection.find() 
for index, record in enumerate(find_all_records): 
    print(f"Record {index}: {record}")
    print("========================================================") 
```

```python
# let's update the data in the collection. 

current_record = {'Name': 'Priya'} 
new_record = {'$set': {'Name': 'Shivani'}}  
```

```python
collection.update_one(current_record, new_record) 
```

```python
# now let's print all records after updating the record. 
find_all_records = collection.find() 
for index, record in enumerate(find_all_records): 
    print(f"Record {index}: {record}")
    print("========================================================")
```

```python
# now we can update multiple records at once. 
present_record = {'City': 'Mumbai', 'City': 'Bangalore'}  
new_record = {'$set': {'City': 'Pune','City': 'Patna'}}
```

```python
collection.update_many(present_record, new_record)
```

```python
for id, record in enumerate(collection.find()):
    print(f"Record {id}: {record}")
    print("========================================================") 
```

```python
# print the limited number of records. 
limited_records = collection.find().limit(5) 

for index, record in enumerate(limited_records):   
    print(f"Record {index}: {record}")
    print("========================================================") 
```

```python
# Now drop the entire collection. 
collection.drop() 
```
